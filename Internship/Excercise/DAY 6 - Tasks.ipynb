# 3. Feature scaling techniques


```python
#  It involves rescaling each feature such that it has a standard deviation of 1 and a mean of 0.
#  to ease the convergence (such as a non-penalized logistic regression), to create a completely different model fit compared to the fit with unscaled data (such as KNeighbors models).
#  Techniques:
##         Absolute Maximum Scaling:
#                     Find the absolute maximum value of the feature in the dataset
#                     Divide all the values in the column by that maximum value

##         Min-Max Scaling:
#                         n min-max you will subtract the minimum value in the dataset with all the values
#                         and then divide this by the range of the dataset(maximum-minimum).
##         Normalization:
                        
#         Standardization
#         Robust Scaling
```

```python
# Overfitting
# They can be unstable, meaning small changes in the data can lead to large changes in the structure of the tree and the predictions

# To overcome:
#     Using pruning techniques, such as reducing the depth or size of the tree, to avoid overfitting and simplify the tree.
#             Post pruning : Find max depth then scale down
#             pre pruning
#     Using ensemble methods, such as random forests or boosting
```

# 2. Gridsearch cv on Random forest dataset

```python
df3 = pd.read_csv("C:/Users/Welcome/Downloads/heart.csv")
df3
```

```python
df3.describe()
```

```python
df3['target'].unique()
```

```python
X = df3.drop(['target'], axis = "columns")
y = df3['target']
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
X_train
```

```python
from sklearn.ensemble import RandomForestClassifier
estimator = RandomForestClassifier()
```

```python
param_grid={
    'criterion':['gini'],
#     "splitter":'best',
    "max_depth":[2,4],
    "min_samples_split":[2,5],
    "min_samples_leaf":[1,3]
    }
param_grid
```

```python
from sklearn.model_selection import GridSearchCV
gscv = GridSearchCV(estimator=estimator,param_grid= param_grid,n_jobs=-1,return_train_score=True)
gscv
```

```python
gscv.fit(X_train, y_train)
```

```python
gscv.best_params_
```

```python
from sklearn import tree
```

```python
tree.plot_tree()
```

```python
gscv.score(X_test,y_test)
```

```python
gscv.score(X_train,y_train)
```

# 3. label encoding in decision tree dataset

```python
df4 = pd.read_csv("C:/Users/Welcome/Downloads/car_evaluation.csv")
df4
```

```python
df4.info()
```

```python
from sklearn.preprocessing import LabelEncoder
le_vhigh = LabelEncoder()
le_vhigh1 = LabelEncoder()
le_2 = LabelEncoder()
le_21= LabelEncoder()
le_small = LabelEncoder()
le_low = LabelEncoder()
le_unacc = LabelEncoder()
```

```python
df4['vhigh_n'] = le_vhigh.fit_transform(df4["vhigh"])
df4['vhigh1_n'] = le_vhigh1.fit_transform(df4["vhigh.1"])
df4['v2_n'] = le_2.fit_transform(df4["2"])
df4['v21_n'] = le_21.fit_transform(df4["2.1"])
df4['small_n'] = le_small.fit_transform(df4["small"])
df4['low_n'] = le_low.fit_transform(df4["low"])
df4['unnac_n'] = le_unacc.fit_transform(df4["unacc"])
```
